```python
import requests
import pandas as pd
import yaml

import sys
sys.path.append(r"D:/aviation_data_pipeline")
import os
os.chdir(r"D:\aviation_data_pipeline")

from plugins.etl.transform import transform_raw_data

# =========================
# CONFIG LOADING
# =========================


# =========================
# URLS (EDIT HERE IF NEEDED)
# =========================
BASE_URL = "https://opensky-network.org/api/states/all"

LA_MIN = 45.8389
LO_MIN = 5.9962
LA_MAX = 47.8229
LO_MAX = 10.5226

URL_FULL = BASE_URL
URL_FILTERED = f"{BASE_URL}?lamin={LA_MIN}&lomin={LO_MIN}&lamax={LA_MAX}&lomax={LO_MAX}"


# =========================
# CORE TEST FUNCTION
# =========================
def run_opensky_test(url, name="OpenSky Test"):
    print("\n" + "=" * 80)
    print(f"🔍 {name}")
    print(f"URL: {url}")
    print("=" * 80)

    # Fetch API
    response = requests.get(url)
    print("\nSTATUS:", response.status_code)

    # Parse JSON
    data = response.json()

    # Transform
    df = transform_raw_data(data)

    # Output preview (Jupyter-friendly if used in notebook)
    try:
        from IPython.display import display
        display(df.head())
    except ImportError:
        print(df.head())

    print("\n📊 Shape:", df.shape)
    print("📊 Columns:", df.columns)

    # Diagnostics
    print("\n🧪 Diagnostics:")
    print("Empty DF:", df.empty)
    print("Total NaNs:", df.isna().sum().sum())

    return df


# =========================
# RUN TESTS
# =========================
if __name__ == "__main__":
    df_full = run_opensky_test(URL_FULL, "FULL DATA")
    df_filtered = run_opensky_test(URL_FILTERED, "FILTERED DATA")
```
